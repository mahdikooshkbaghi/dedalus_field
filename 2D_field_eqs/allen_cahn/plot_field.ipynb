```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
# matplotlib setup
publication_dict = {
    "text.usetex": True,                # use LaTeX to write all text
    "font.family": "serif",
    "axes.labelsize": 12,               # LaTeX default is 10pt font.
    "font.size": 12,
    "legend.fontsize": 12,               # Make the legend/label fonts a little smaller
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "image.cmap": 'Spectral',
}
matplotlib.rcParams.update(publication_dict)
%config InlineBackend.figure_format = 'retina'
```

```python
timestep = 1e-5
saved_every = 10
tt = timestep*saved_every
u = np.load('u.npy')
print(f'shape of the field array is {u.shape}')
```

```python
fig, axs = plt.subplots(5,4, figsize=(10, 8))
axs_r = axs.ravel()

t_range = np.linspace(0, u.shape[0], num=20)
for i, t in enumerate(t_range):
    axs_r[i].imshow(u[i,:,:])
    axs_r[i].set_title(f't={tt*t:.4f}')
plt.tight_layout()
plt.savefig('allen_cahn.png', dpi=300)
```
